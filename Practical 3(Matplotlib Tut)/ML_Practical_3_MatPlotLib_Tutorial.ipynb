# Markers

**Mark each point with a circle**:

```python
import matplotlib.pyplot as plt
import numpy as np
ypoints = np.array([3, 8, 20, 43])
#plot 1
plt.plot(ypoints, marker = 'X')
plt.show()

#plot2
plt.plot(ypoints, 'o:r') # circle, dotted line, red
plt.show()
```

# Marker Size

**Set the size of the markers to 20:**

```python
import matplotlib.pyplot as plt
import numpy as np
ypoints = np.array([3, 8, 20, 43])
plt.plot(ypoints, marker = 'X', ms = 20)
plt.show()
```

# Marker Color

**Set the EDGE color to red**:

```python
import matplotlib.pyplot as plt
import numpy as np
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'X', ms = 20, mec = 'r')
plt.show()
```

**Set the FACE color to red:**

```python
import matplotlib.pyplot as plt
import numpy as np
ypoints = np.array([3, 8, 20, 43])
plt.plot(ypoints, marker = 'X', ms = 20, mfc = 'r')
plt.show()
```

# Linestyle

# Multiple Lines

**Draw two lines by specifying a plt.plot() function for each line:**

```python
import matplotlib.pyplot as plt
import numpy as np
y1 = np.array([3, 8, 20, 43])
y2 = np.array([6, 2, 43, 20])
plt.plot(y1,color="red")
plt.plot(y2,color="green")
plt.show()
```

**Draw two lines by specifying the x- and y-point values for both lines:**

```python
import matplotlib.pyplot as plt
import numpy as np
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 20, 1, 8])
x2 = np.array([0, 7, 2, 3])
y2 = np.array([6, 2, 7, 43])
plt.plot(x1, y1, x2, y2)
plt.show()
```

**bold text**# Create Labels for a Plot

**Add labels to the x- and y-axis:**

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([143, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 134, 223, 123, 280, 210, 30, 310, 130, 160])
plt.plot(x, y)
plt.title("Average Play of AB DeVilliers", loc = 'left')
plt.xlabel("Scored Runs")
plt.ylabel("Strike Rate")
plt.show()
```

# Set Font Properties for Title and Labels

**Set font properties for the title and labels:**

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([143, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 134, 223, 123, 280, 210, 30, 310, 130, 160])
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
plt.title("Average Play of AB DeVilliers", fontdict = font1)
plt.xlabel("Scored Runs", fontdict = font2)
plt.ylabel("Strike Rate", fontdict = font2)
plt.plot(x, y)
plt.show()
```

# Add Grid Lines to a Plot

**Add grid lines to the plot:**

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([80, 85, 90, 95, 150, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 230, 280, 290, 300, 310, 320, 330])
plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.plot(x, y)
plt.grid(color="red")
plt.show()
```

# The subplots() Function

**Draw 2 plots on top of each other:**

```python
import matplotlib.pyplot as plt
import numpy as np
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 20, 43])
plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("SALES")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("INCOME")
plt.show()
```

# Creating Bars

**Draw 4 bars:**

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.bar(x, y, color = "red")
plt.show()
```

## Bar Width and Height

Implementation

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.bar(x, y, width = 0.1)
plt.show()
```

# Create Histogram

**A simple pie chart:**

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array([70, 44, 37, 39])
mylabels = ["Virat Kohli", "Joe Root", "Kane Williamson", "Steve Smith"]
plt.pie(y,labels = mylabels)
plt.show()
```

# Start Angle

**Start the first wedge at 90 degrees:**

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array([70, 44, 37, 39])
mylabels = ["Virat Kohli", "Joe Root", "Kane Williamson", "Steve Smith"]
plt.pie(y, labels = mylabels, startangle = 90)
plt.show()
```

# Explode

**Pull the "Virat Kohli" wedge 0.2 from the center of the pie:**

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array([70, 44, 37, 39])
mylabels = ["Virat Kohli", "Joe Root", "Kane Williamson", "Steve Smith"]
myexplode = [0.2, 0, 0, 0]
plt.pie(y, labels = mylabels, explode = myexplode,shadow = True)

plt.show()
```
