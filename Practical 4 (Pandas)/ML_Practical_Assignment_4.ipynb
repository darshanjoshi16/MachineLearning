Create a custom dataframe and apply logical, query and filter methods on the
same

```python
df = pd.DataFrame({'name':['Jane','John','Ashley','Mike','Emily','Jack','Catlin'],
'ctg':['A','A','C','B','B','C','B'],
'val':np.random.random(7).round(2),
'val2':np.random.randint(1,10, size=7)
})

names = ['John','Catlin','Mike']

print("============================\n")
print("Initial Dataframe\n")
print(df)

print("============================\n")
print("Applying the logical OR operator\n")
print(df[df.val > 0.5])

print("============================\n")
print("Applying the logical > operator\n")
print(df[df.name > 'Jane'])

print("============================\n")
print("Working with strings using isin method\n")
print(df[df.name.isin(names)])

print("============================\n")
print("Usage of tilde sign as a NOT operator\n")
print(df[~df.name.str.startswith('J')])

print("============================\n")
print("Usage of Query Method\n")
print(df.query('ctg == "B" and val > 0.5'))
```

Calculate the mean, median and mode for the following dataframe.

```python
dataMatrix = {"D1":[135, 137, 136, 138, 138],
"D2":[43, 42, 42, 42, 42],
"D3":[72, 73, 72, 72, 73],
"D4":[100, 102, 100, 103, 104] };

dataFrame = pd.DataFrame(data=dataMatrix);
print("DataFrame:");

print("Mean:Computed column-wise:");
meanData = dataFrame.mean(axis=1);
print(meanData);

print("Mean:Computed row-wise:");
meanData = dataFrame.mean(axis=0);
print(meanData);
```

```python
print("Median:Computed column-wise:");
medianData = dataFrame.median(axis=1);
print(medianData);

print("Median:Computed row-wise:");
medianData = dataFrame.median(axis=0);
print(medianData);
```

```python
print("Mode:Computed column-wise:");
modeData = dataFrame.mode(axis=1);
print(modeData);

print("Mode:Computed row-wise:");
modeData = dataFrame.mode(axis=0);
print(modeData);
```
