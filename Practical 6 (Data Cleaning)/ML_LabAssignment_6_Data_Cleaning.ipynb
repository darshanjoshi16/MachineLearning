```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv")
df
```

7.  Fill NaN values of each column with either of the following methods:

a. Fill Missing DataFrame Values with Column Mean, Median or Mode

b. Fill Missing DataFrame Values with a Constant

c. Forward Fill/Backward Fill Missing DataFrame Values

```python
#Using mode
df['boat'].fillna(df['boat'].mode(), inplace=True)
df

# using the global constant
df['boat'].fillna(5, inplace=True)
df

#forward fill method
df['body'].fillna(method='ffill', inplace=True)
df
```

10. **Identify the outliers and remove the outliers.**

```python
Q1 = df.quantile(0.25) #for each column
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)

lower=Q1-1.5 * IQR
print("------lOWER Range-----")
print(lower)

print("------Upper Range-----")
upper=Q3 + 1.5 * IQR
print(upper)
```

```python
df_outliers = df[((df < lower) |(df > upper)).any(axis=1)]
print(df_outliers.shape)
df_out_final = df[~((df < lower) |(df > upper)).any(axis=1)] #removing outlier...remaining data
df_out_final.shape
```
